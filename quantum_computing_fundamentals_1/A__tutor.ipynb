# PART 1: QUANTUM INTERFERENCE

## What is Quantum Interference?

**In my own words:**

Quantum interference is when quantum states combine so that some results become stronger and some cancel out.  
This helps a quantum computer hide wrong answers and make the correct answer stand out.

**Step 1: First H gate does...**

# PART 2: DEUTSCH-JOZSA ALGORITHM

## The Problem It Solves

**What is the mystery box problem?**

We are given a function that is either constant or balanced.  
We must figure out which type it is.

**Constant function means:**

The output is always 0 or always 1.

**Balanced function means:**

Half the outputs are 0 and half are 1.

---

## Classical vs Quantum Solution

### Classical Computer Approach

**How many checks does a classical computer need?**

In the worst case, it needs more than half of all inputs.

**Why can't it do better?**

Because it checks inputs one at a time.

---

### Quantum Computer Approach

**How many checks does a quantum computer need?**

Only one.

**How is this possible?**

It uses superposition and interference to test all inputs at once.

---

## How the Circuit Works

**Step 1:**  
Apply H gates to put qubits into superposition.

**Step 2:**  
The oracle changes the phase depending on the function.

**Step 3:**  
Apply H gates again to create interference.

**Step 4:**  
Measure the qubits.
